```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def comparar_tiempos(archivo_dinamico, archivo_bruto,titulo):
    # Cargar los datos
    df_dinamico = pd.read_csv(archivo_dinamico)
    df_bruto = pd.read_csv(archivo_bruto)
    
    # Calcular el largo máximo de los strings
    df_dinamico['largo_max'] = df_dinamico[['largo_str1', 'largo_str2']].max(axis=1)
    df_bruto['largo_max'] = df_bruto[['largo_str1', 'largo_str2']].max(axis=1)

    # Aplicar logaritmo a los tiempos de ejecución para mejor visualización
    df_dinamico['log_tiempo'] = np.log10(df_dinamico['tiempo_microsegundos'] + 1)
    df_bruto['log_tiempo'] = np.log10(df_bruto['tiempo_microsegundos'] + 1)

    # Crear el gráfico de dispersión
    plt.figure(figsize=(10, 6))
    
    # Gráfico para datos dinámicos
    plt.scatter(df_dinamico['largo_max'], df_dinamico['log_tiempo'], label='Programación Dinámica', alpha=0.7)
    
    # Gráfico para datos de fuerza bruta
    plt.scatter(df_bruto['largo_max'], df_bruto['log_tiempo'], label='Fuerza Bruta', alpha=0.7)
    
    # Configurar el gráfico
    plt.xlabel('Largo Máximo de los Strings')
    plt.ylabel('Log(Tiempo de Ejecución) (Microsegundos)')
    plt.title(titulo)
    plt.legend()
    plt.grid(True)
    plt.show()
```
